```python
#Define action space
A = ["U", "D", "L", "R"]

#Define transition probabilities
p = 0.02
anticipated_p = 1-p
neighbor_p = p/3
```

```python
#Define reward function
def reward(state):
  i, j = state
  cell = maze[i][j]
  if cell == 1:  # wall
      return None
  elif cell == 2 or cell == 3:  # oil or bump
      return -5 if cell == 2 else -10
  elif cell == 5:  # goal
      return 200
  else:  # empty space
      return -1
```

```python
#Initialize policy all left and value states zero
policy= {s: "L" for s in S}
V= {s: 0 for s in S}
```

Policy iteration: Base case scenario

```python
gamma = 0.95
theta = 0.01

# Policy iteration
while True:
    # Policy evaluation
    while True:
        delta = 0
        for s in S:
            v = V[s]
            initial_action = policy[s]
            next_states = [(s[0]-1, s[1]), (s[0]+1, s[1]), (s[0], s[1]-1), (s[0], s[1]+1)]
            next_values = []
            for i, next_state in enumerate(next_states):
                if next_state in S:
                    anticipated_value = anticipated_p * (reward(next_state) + gamma * V[next_state])
                    neighbor_value = 0
                    for j, neighbor_state in enumerate(next_states):
                        if i != j and neighbor_state in S:
                            neighbor_value += neighbor_p * (reward(neighbor_state) + gamma * V[neighbor_state])
                    next_value = anticipated_value + neighbor_value
                    next_values.append(next_value)
                else:
                    next_values.append(-np.inf)
            V[s] = np.max(next_values)
            delta = max(delta, abs(v - V[s]))
        if delta < theta:
            break

    # Policy improvement
    policy_stable = True
    for s in S:
        initial_action = policy[s]
        next_states = [(s[0]-1, s[1]), (s[0]+1, s[1]), (s[0], s[1]-1), (s[0], s[1]+1)]
        action_values = []
        for i, next_state in enumerate(next_states):
            if next_state in S:
                anticipated_value = anticipated_p * (reward(next_state) + gamma * V[next_state])
                neighbor_value = 0
                for j, neighbor_state in enumerate(next_states):
                    if i != j and neighbor_state in S:
                        neighbor_value += neighbor_p * (reward(neighbor_state) + gamma * V[neighbor_state])
                next_value = anticipated_value + neighbor_value
                action_values.append(next_value)
            else:
                action_values.append(-np.inf)
        best_action = A[np.argmax(action_values)]
        policy[s] = best_action
        if initial_action != best_action:
            policy_stable = False
    if policy_stable:
        break


```

# **Part 2: Value Iteration Method**

```python
# Initialize state values to be all zeros
V = {s: 0 for s in S}

gamma = 0.95
theta = 0.01
```

```python
def next_state(state, action):
    if action == "U":
        return (state[0] - 1, state[1])
    elif action == "D":
        return (state[0] + 1, state[1])
    elif action == "L":
        return (state[0], state[1] - 1)
    elif action == "R":
        return (state[0], state[1] + 1)

def action_value(state, action, V):
    next_s = next_state(state, action)
    if next_s in S:
        anticipated_value = anticipated_p * (reward(next_s) + gamma * V[next_s])
        neighbor_value = 0
        for neighbor_action in A:
            if neighbor_action != action:
                neighbor_s = next_state(state, neighbor_action)
                if neighbor_s in S:
                    neighbor_value += neighbor_p * (reward(neighbor_s) + gamma * V[neighbor_s])
        return anticipated_value + neighbor_value
    else:
        return -np.inf
```

```python
# Value iteration
while True:
    delta = 0
    for s in S:
        v = V[s]
        action_values = [action_value(s, a, V) for a in A]
        V[s] = np.max(action_values)
        delta = max(delta, abs(v - V[s]))
    if delta < theta:
        break

# Define policy based on the state values
policy = {s: None for s in S}
for s in S:
    action_values = [action_value(s, a, V) for a in A]
    best_action = A[np.argmax(action_values)]
    policy[s] = best_action
```

## **Policy iteration vs Value iteration for this problem**

Policy iteration and value iteration are two different algorithms for solving Markov Decision Processes (MDPs). They both aim to find the optimal policy (the best set of actions to take in each state) and the optimal state values (the expected return from each state). However, they differ in their approach.
Policy Iteration consists of two main steps:
1. Policy Evaluation: Given a policy, it calculates the state values by iteratively updating the state values according to the Bellman equation for the current policy.
2. Policy Improvement: Given the state values, the policy is updated by selecting the best action for each state based on the maximum expected return.
These two steps are alternated until the policy converges to the optimal policy.


Value Iteration combines the policy evaluation and policy improvement steps into a single loop. It calculates the state values by iteratively updating them according to the Bellman optimality equation, which considers the best action for each state in the update. When the state values converge, the optimal policy can be extracted from the final state values.


# **For our scenario, we saw that value iteration took more time to execute, that is it took more iterations than policy iteration method. The results obtained are exactly the same for both the alogirthms, they both give the same optimal policy and state value functions. But since value iteration does the computation in a single loop, it uses less resources. Policy iteration uses more resources but converges faster. Thus depending on the requirement and scenario, we can decide which algorithm to use.**
